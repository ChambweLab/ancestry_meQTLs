# eQTM community detection in the NCI-Maryland cohort

## Import libraries

```python
import networkx as nx
import pandas as pd
```

```python
print(nx.__version__)
```

Import data

```python
df = pd.read_csv('maryland_AFR-sig_eQTMs_pval.csv')
```

## Create network based on eQTM results

```python
net = nx.from_pandas_edgelist(df, source='probe', target='gene', edge_attr="rho") #edge_attr=True
```

Save weights as separate list

```python
weights = [i['rho'] for i in dict(net.edges).values()]
```

## Detect greedy communities within the network

```python
G = nx.Graph(net)
```

```python
from networkx.algorithms.community import greedy_modularity_communities
comm = greedy_modularity_communities(G)
```

```python
print(f"The eQTM network has {len(comm)} communities.")
```

```python
test = [list(x) for x in comm]
```

```python
df = pd.DataFrame(list(zip(test)))
```

Save communities as csv

```python
df.to_csv("AFR-all_comm_maryland.csv")
```

# Clean-up gnomad frequency info to obtain AFR and EUR specific AFs for meSNPs

## Load libraries

```r
library(readr)
library(data.table)
library(stringr)
```

## Obtain population-specific AF from gnomad variant annotation table AF entry

Load AF info for meSNPs obtained when filtering the gnomad HailTable

```r
dat1 <- read.csv("freq_gnomad_meSNPs.tsv",sep="")
head(dat1)
dim(dat1)
```

Load gnomad frequency globals (information on AF HailTable contents) and process data accordingly

```r
globals <- read.csv("freq_globals.tsv",sep="")
head(globals)
```

The information stored in the AF metadata is informing as for the order that the population specific AFs appear in the frequency column. More specifically, the first 12 sequencies from gnomad AF entry represent the raw and adjusted allele frequency in a global scale, as well as adjusted AFs for 9 different population groups and an additional group called  "remaining individuals" group.

```r
proc <- dat1
proc$X__uid_5 <- stringr::str_extract(proc$X__uid_5, "^([^,]+,){12}") # Extract first 12 entries with AFs of interest 
```

Remove "[" and last comma

```r
proc$X__uid_5 <- gsub("\\[","",proc$X__uid_5)
proc$X__uid_5 <-  gsub(",$", "", proc$X__uid_5)
```

Seperate AFs for the different populations

```r
column_names = c("global_adj","global_raw","nfe_adj","fin_adj","mid_adj","remaining_adj",
                "ami_adj","asj_adj","afr_adj","eas_adj","sas_adj","amr_adj")

final <- separate(data = proc, col = X__uid_5, into = column_names, sep = ",")
```

## Save results

```r
write.csv(final,"gnomad_vat_table_processed_MULTI_allele.csv",row.names=F)
```
